# GIỚI THIỆU DỰ ÁN

- Dự án Báo cáo thường niên về nhà phát triển Stack Overflow là một nghiên cứu quy mô lớn được thực hiện hàng năm để khảo sát và thu thập thông tin về cộng đồng nhà phát triển trên toàn cầu. Báo cáo này cung cấp cái nhìn sâu sắc về các xu hướng, quan điểm, ngôn ngữ lập trình, công nghệ và các vấn đề quan trọng khác đang ảnh hưởng đến lĩnh vực phát triển phần mềm.
- Tháng 5/2022, hơn 70.000 nhà phát triển đã chia sẻ với StackOverflow về cách họ học hỏi và thăng cấp, những công cụ họ đang sử dụng và những gì họ muốn.
- Dự án phân tích Báo cáo thường niên của Stack Overflow nhằm mục tiêu đưa ra thông tin và insight hữu ích từ dữ liệu khảo sát giúp hiểu rõ về cộng đồng nhà phát triển, xu hướng công nghệ và các vấn đề liên quan đến phát triển phần mềm

Khảo sát tập trung vào các chủ đề:
1. Hồ sơ nhà phát triển/ Hồ sơ người trả lời (Developer Profile/ Respondent's profile):
- Education
- Learning to code
- Experience
- Developer roles
- Key territories
- Demographics

2. Công nghệ (Technology): Các công cụ và công nghệ mà các nhà phát triển hiện đang sử dụng và những gì họ muốn sử dụng.

3. Công việc (Work): Việc làm của người trả lời, thông tin công ty, tiền lương, tầm ảnh hưởng, sở thích, v.v

4. Cộng đồng (Community): Cách mọi người sử dụng Stack Overflow và mức độ kết nối của họ với cộng đồng StackOverflow.

5. Nhà phát triển chuyên nghiệp (Professional Developers) :Lối sống làm việc, kinh nghiệm làm việc của Người trả lời, v.v.

6. Phương pháp (Methodology): Phản hồi về khảo sát.

```python
def transform_list(list, empl_dict):
    transformed_list  = []
    for item in list:
        transformed_item = ' '.join([empl_dict[i] for i in item])
        transformed_list.append(transformed_item)
    return transformed_list
```

## Which programming, scripting, and markup languages have you done extensive development work in over the past year


```python
data['languagehaveworkedwith'].value_counts()
```

```python
language = data['languagehaveworkedwith'].str.split(';')
```

```python
language
```

```python
sub_language = []
for i in language:
    sub_language.extend(i)
```

```python
sub_language
```

```python
language_list = ['Java','JavaScript','Python','SQL','C#','PHP','HTML/CSS','R','Swift','C++','Assembly']
```

```python
language_dict = {}
for i in sub_language:
    for j in language_list:
        if j in i:
            language_dict[j] = language_dict.get(j,0) + 1 # Nếu giá trị j xuất hiện trong giá trị i, language_dict[j] được tăng lên 1. Phương thức get(j, 0) trả về giá trị hiện tại của language_dict[j] nếu tồn tại, nếu không, nó trả về 0.
```

```python
language_dict
```

```python
language_index = []
language_value = []
for i, v in language_dict.items():
    language_index.append(i)
    language_value.append(v)
```

```python
plt.barh(language_index,language_value, color = '#ff5757')
plt.xlabel('Số lượng người dùng')
plt.show()
```

Từ đồ thị trên, có thể thấy được rằng Java, JavaScript, SQL, HTML/CSS là những ngôn ngữ lập trình được nhiều người sử dụng nhất
